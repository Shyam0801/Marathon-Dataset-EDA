## Filtering only the USA Data (only 50km and 50mi event distance & of year 2020)

```python
#Getting only the USA Event Data Separtely
df_usa = df[(df['Event distance/length'].isin(['50km', '50mi'])) & 
   (df['Year of event'] == 2020) & 
   (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
#making the change in the data frame df_usa
df_usa['Event name'] = df_usa['Event name'].str.split('(').str.get(0)
```

```python
#Checking the null values in the Athlete Year of Birth Columns as the conversion to integer failed
df_usa[df_usa['Athlete year of birth'].isna()]
```

```python
#Changing the Datatype of Athlete year of birth to INT
df_usa['Athlete year of birth'] = df_usa['Athlete year of birth'].astype(int)
df_usa.head()
```

```python
#Calculating the Athlete age (as the event happened in 2020, we need to calculate the athlete age at that time)
df_usa['Athlete Age'] = 2020 - df_usa['Athlete year of birth']
df_usa.head()
```

```python
#Removing the trailing h from the Athlete performance Column
df_usa['Athlete performance'] = df_usa['Athlete performance'].str.split(' ').str.get(0)
```

```python
#Checking the Data to make sure it has changed
df_usa.head()
```

```python
#Dropping some of the Columns which are not necessary
df_usa.drop(['Athlete club', 'Athlete country', 'Athlete age category', 'Athlete year of birth'], axis = 1, inplace = True)
df_usa.head()
```

```python
#Changing the Datatype of Athlete average speed to float
df_usa['Athlete average speed'] = df_usa['Athlete average speed'].astype(float)
df_usa.dtypes
```

```python
#standardizing the column names

#changing all to lower case first
df_usa.columns = df_usa.columns.str.lower()

#Changing the columns space to _
df_usa.columns = df_usa.columns.str.replace(' ', '_')

#Checkign the datset now to ensure it  has changed
df_usa.head()
```

```python
#Creating the event_month column from the event_date to group them by seasons

#Checking the dates
df_usa['event_dates'].value_counts()
```

```python
#As there are some dates like 28.-29.11.2020 , we can't simply split and get the month, it will not work
#So, let's first split the date column and see 
df_usa['event_dates'].str.split('-').value_counts()
```

```python
#Splitting the event_dates into event_start date and event_last_date
df_usa[['event_start_date', 'event_end_date']] = df_usa['event_dates'].str.split('-', expand = True)
df_usa.head()
```

```python
# Now finding out where the event_start_date is less than the lenth of 4
df_usa[df_usa['event_start_date'].apply(len) < 4]
```

```python
#We need to fill the event_start_date with the month and year from event_end_date, 
                                #where the event_start_date length is less than 4
#First, getting the month and year from the end_date

df_usa['event_end_date'].str[3:].value_counts()
```

```python
# Now Making another new columns as final_event_start_date, by combining above two formulas
df_usa['final_event_start_date'] = df_usa.apply(
        lambda x: x['event_start_date'] + x['event_end_date'][3:] if len(x['event_start_date']) < 4
                                                                else x['event_start_date'], axis = 1)

df_usa.head()
```

```python
#now checking the event_start_date with the length of less than 4, and checking the new column to see it is changed
df_usa[df_usa['event_start_date'].apply(len) < 4]
```

```python
#now dropping the event_start_date, event_end_date
df_usa.drop(columns = ['event_start_date', 'event_end_date'], inplace = True)
```

```python
#now renaming the final_event_start_date to event_date_corrected
df_usa = df_usa.rename(columns = {'final_event_start_date' : 'event_date_corrected'})
df_usa.head()
```

```python
#Now Separating the Month from the event_date_corrected
df_usa['event_month'] = df_usa['event_date_corrected'].str.split('.').str.get(1)
df_usa.head()
```

```python
#Changing the type as int here for event_month
df_usa['event_month'] = df_usa['event_month'].astype(int)
df_usa.head()
```

```python
#Dropping the column event_date_corrected from the dataframe
df_usa.drop(columns = ['event_date_corrected'], inplace = True)
```

```python
#Checking if it is dropped
df_usa.head()
```

```python
#Spring 3 - 5
#Summer 6 - 8
#Fall 9 - 11
#Winter 12 - 2

#Now creating another column as event_season
df_usa['event_season'] = df_usa['event_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
#Checking if it worked
df_usa.head()
```

```python
#Checking athlete's average speed in each season and count of athletes run in every season
df_usa.groupby('event_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
